```python
import json

def rank_players():
    # Load the master player database
    try:
        with open('../src/data/players/master_player_database.json', 'r') as f:
            database = json.load(f)
    except FileNotFoundError:
        print('Error: master_player_database.json not found. Make sure the path is correct.')
        return

    players = database.get('players', [])
    
    batsmen = []
    bowlers = []

    for player in players:
        # Batsmen ranking (includes batsmen, wicket-keepers, and all-rounders)
        if player.get('primaryPlaystyle', {}).get('batting'):
            primary_playstyle = player['primaryPlaystyle']['batting']
            top_playstyle = player.get('topPlaystyles', {}).get('batting', [])
            if top_playstyle and top_playstyle[0]['name'] == primary_playstyle:
                rating = top_playstyle[0]['rating']
                batsmen.append({
                    'name': player['name'],
                    'role': player['role'],
                    'playstyle': primary_playstyle,
                    'rating': rating
                })

        # Bowlers ranking (includes bowlers and all-rounders)
        if player.get('primaryPlaystyle', {}).get('bowling'):
            primary_playstyle = player['primaryPlaystyle']['bowling']
            top_playstyle = player.get('topPlaystyles', {}).get('bowling', [])
            if top_playstyle and top_playstyle[0]['name'] == primary_playstyle:
                rating = top_playstyle[0]['rating']
                bowlers.append({
                    'name': player['name'],
                    'role': player['role'],
                    'playstyle': primary_playstyle,
                    'rating': rating
                })

    # Sort the lists by rating in descending order
    batsmen.sort(key=lambda x: x['rating'], reverse=True)
    bowlers.sort(key=lambda x: x['rating'], reverse=True)

    # Print the ranked lists
    print('--- Top Batsmen (Ranked by Primary Playstyle Rating) ---')
    for i, p in enumerate(batsmen, 1):
        print(f"{i}. {p['name']} ({p['role']}) - Playstyle: {p['playstyle']}, Rating: {p['rating']:.2f}")

    print('--- Top Bowlers (Ranked by Primary Playstyle Rating) ---')
    for i, p in enumerate(bowlers, 1):
        print(f"{i}. {p['name']} ({p['role']}) - Playstyle: {p['playstyle']}, Rating: {p['rating']:.2f}")

rank_players()
```
